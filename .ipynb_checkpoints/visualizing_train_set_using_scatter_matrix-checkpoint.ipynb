```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
df["income_cat"]=pd.cut(df["median_income"], bins =[0,3,5,6,8,np.inf],labels=[1,2,3,4,5])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set,test_set=train_test_split(df,test_size=0.2,stratify=df["income_cat"],random_state=42)
```

```python
train_set
```

```python
test_set
```

```python
for sett in (train_set,test_set):
    sett.drop("income_cat",inplace=True, axis=1)
    
 
```

```python
df.drop("ocean_proximity",inplace=True, axis=1)
```
